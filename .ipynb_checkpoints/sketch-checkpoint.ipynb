```python
import woodruffstats.auth as auth
import woodruffstats.read_credentials as read_credentials
import woodruffstats.get_data as get_data
import woodruffstats.helpers as helpers
import requests
from datetime import datetime
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
credentials = read_credentials.read_credentials()
client_id = credentials['key']
client_secret = credentials['secret']
access_token = auth.get_access_token(credentials)

auth_headers = auth.construct_auth_headers(client_id, access_token)
```

```python
print(access_token)
```

```python
user_id = auth.get_user_id(access_token)
```

```python
workouts = get_data.get_all_workouts(auth_headers, user_id)
```

```python
runs = get_data.get_workouts_runs(workouts)
```

```python
runs_df = pd.DataFrame([helpers.flatten(run) for run in runs])

runs_df['start_datetime_obj'] = [helpers.get_datetime(i) for i in runs_df['start_datetime']]
runs_df['start_datetime_year'] = [i.year for i in runs_df['start_datetime_obj']]
runs_df['start_datetime_month'] = [i.month for i in runs_df['start_datetime_obj']]
runs_df['start_datetime_week'] = [i.week for i in runs_df['start_datetime_obj']]
```

```python
run_distance_by_year = runs_df.groupby('start_datetime_year').sum()['aggregates_distance_total']
data = pd.DataFrame(helpers.convert_metres_to_miles(runs_by_year)).reset_index()
data.columns = ['year', 'distance_miles']

sns.set_style("whitegrid")
g = sns.barplot(data=data, x='year', y='distance_miles', color = '#33B5FF')

for i, v in enumerate(data['distance_miles']):
    g.text(i - 0.31, v - 50, str(round(v, 1)), color='white', fontweight='bold')
```

```python
run_distance_by_year_month = runs_df.groupby(['start_datetime_year', 'start_datetime_month']).sum()['aggregates_distance_total']
data_1 = pd.DataFrame(helpers.convert_metres_to_miles(run_distance_by_year_month)).reset_index()
data_1['year_month'] = [str(i[0]) + '-' + str(i[1]) for i in zip(data_1['start_datetime_year'], data_1['start_datetime_month'])] 
data_1.columns = ['year', 'month', 'distance_miles', 'year_month']
data_1


sns.barplot(data=data_1, x='year_month', y='distance_miles')
```

```python
run_distance_by_year_week = runs_df.groupby(['start_datetime_year', 'start_datetime_week']).sum()['aggregates_distance_total']
data_2 = pd.DataFrame(helpers.convert_metres_to_miles(run_distance_by_year_week)).reset_index()
data_2['year_week'] = [str(i[0]) + '-' + str(i[1]) for i in zip(data_2['start_datetime_year'], data_2['start_datetime_week'])] 
data_2.columns = ['year', 'week', 'distance_miles', 'year_week']
data_2


sns.barplot(data=data_2, x='year_week', y='distance_miles')
```
